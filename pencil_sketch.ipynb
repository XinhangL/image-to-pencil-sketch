```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
img = cv2.imread("dog.jpg")

# Check if the image was loaded successfully
if img is None:
    raise FileNotFoundError(" Couldn't load image. Make sure dog.jpg is in the same folder as this notebook.")

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Invert the grayscale image
inverted = 255 - gray

# Apply Gaussian Blur
blurred = cv2.GaussianBlur(inverted, (61, 61), 0)

# Invert the blurred image
inverted_blurred = 255 - blurred

# Create the sketch
sketch = cv2.divide(gray, inverted_blurred, scale=256.0)

# Display the result
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Pencil Sketch")
plt.imshow(sketch, cmap='gray')
plt.axis("off")

plt.show()

# Optional: Save the result
cv2.imwrite("sketch_result.jpg", sketch)
print(" Sketch saved as sketch_result.jpg")
```
